```python
request = "何人かの LLM に彼らの知性を評価するために尋ねることができる、挑戦的で微妙な質問を考え出してください。"
request += "質問のみに答え、説明は不要です。"
messages = [{"role": "user", "content": request}]
```

```python
openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
)
question = response.choices[0].message.content
print(question)
```
